```python
import numpy as np
import pandas as pd
import os
import gzip
import scipy.misc
from PIL import Image
import matplotlib.cm as cm
from matplotlib import pyplot as plt
from scipy import optimize

data_path = ""
```

```python
def mnsit_to_csv(root, imgf, labelf, outf, n):
    
    imgf = os.path.join(root, imgf)
    labelf = os.path.join(root, labelf)
    outf = os.path.join(root, outf)
    
    with gzip.open(imgf, "rb") as f:
        with gzip.open(labelf, "rb") as l:
            f.read(16)
            l.read(8)
            images = []

            for i in range(n):
                image = [ord(l.read(1))]
                for j in range(28*28):
                    image.append(ord(f.read(1)))
                images.append(image)

            with open(outf, "w") as o:
                for image in images:
                    o.write(",".join(str(pix) for pix in image)+"\n")
            print "{} lines written to file {}".format(len(images), outf)
```

```python
mnsit_to_csv(data_path, "test/t10k-images-idx3-ubyte.gz", "test/t10k-labels-idx1-ubyte.gz","mnist_test.csv", 10000)
```

```python
def get_train_test(root, train_file, test_file):
    train_path = os.path.join(root, train_file)
    test_path = os.path.join(root, test_file)
    
    train = np.loadtxt(train_path, delimiter=',')
    test = np.loadtxt(test_path, delimiter=',')
    
    return train, test
    
```

```python
#Defining Cost function for logistic regression
def h(theta, X):
    return 1/(1+np.exp(-np.dot(X,theta)))

def computeCost(theta, X, y, lambda_=0):
    m = y.shape[0]
    h_ = h(theta, X)
    term_1 = -np.dot(y, np.log(h_))
    term_2 = -np.dot(1-y, np.log(1-h_))
    return (term_1+term_2)/m
    
def computeGradient(theta, X, y, lambda_=0):
    m = y.shape[0]
    print y.shape
    h_ = h(theta, X)
    print h_.shape
    print X.shape
    return np.dot(h_-y, X)/m
    
y = train[:,0]
X = train[:,1:]
theta = np.zeros(X.shape[1])
lambda_ = 0
```

```python
def optimize_theta(theta, X, y, lambda_):
    for i in range(10):
        i_ = 10 if i is 0 else i 
        y_ = np.array([1 if e==i else 0 for e in y])
        
        Theta = np.zeros((10,X.shape[1]))
        print 'Optimizing for {}'.format(i)
        result = optimize.fmin_cg(computeCost, fprime=computeGradient, x0=theta, args=(X, y_, lambda_), maxiter=50,disp=False, full_output=True)
        Theta[i,:] = result[0]
    return Theta
```
